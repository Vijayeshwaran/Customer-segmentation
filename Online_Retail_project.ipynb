```python
# Chossing the necessary features for the customer segmentation
# We are going to cluster the customers based on Recency, Monetary, Frequency
# Recency - Number of days since the last purchase for each customer.
# Monetary - Total transaction done by each customer (Total Sales).
# Frequency - Number of transaction for each customer.
```

```python
df_copy = df
```

```python
Frequency = df.groupby(['CustomerID'])['InvoiceDate'].count()
Frequency = Frequency.reset_index()
print(Frequency)
```

```python
Monetary = df.groupby(['CustomerID'])['TotalSales'].sum()
Monetary = Monetary.reset_index()
print(Monetary)
```

```python
from datetime import datetime
max_date = max(df['InvoiceDate'])
df['Date_diff'] = max_date - df['InvoiceDate']
Recency = df.groupby(['CustomerID'])['Date_diff'].min()
Recency = Recency.reset_index()
for i in range(len(Recency)):
  Recency['Date_diff'][i] = Recency['Date_diff'][i].days
```

```python
print(Recency)
```

```python
rfm_data = pd.DataFrame(pd.concat([Frequency, Monetary['TotalSales'], Recency['Date_diff']], axis = 1))
rfm_data.rename(columns = {'InvoiceDate' : 'Frequency', 'TotalSales' : 'Monetary', 'Date_diff' : 'Recency'}, inplace = True)
rfm_data = pd.DataFrame(rfm_data)
rfm_data
```

```python
rfm_data_copy = rfm_data
```

```python
# We are going to cluster the customers using K-Means model which is based on
# Euclidean distance, So standardizing is essential.
# Standardizing the rmf_data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(rfm_data[['Frequency', 'Monetary', 'Recency']])
print(scaled_df)
```

```python
# Model Building
from sklearn.cluster import KMeans
# Finding the optimal number of clusters
sse = {}
for i in range(2, 11):
  kmeans = KMeans(n_clusters=i, random_state = 0).fit(scaled_df)
  sse[i] = kmeans.inertia_
print(sse)
```

```python
# Elbow plot
plt.figure(figsize=(10,5))
plt.plot(sse.keys(), sse.values(), marker = 'o')
plt.title('Elbow plot')
plt.xlabel('Clusters')
plt.ylabel('sse')
```

```python
# Silhouette score
from sklearn.metrics import silhouette_score
sc = {}
for i in range(2, 11):
  kmeans = KMeans(n_clusters=i, random_state = 0).fit(scaled_df)
  labels = kmeans.predict(scaled_df)
  sc[i] = silhouette_score(scaled_df, labels)
print(sc)
```

```python
# Silhouette score plot
plt.figure(figsize = (10, 5))
plt.plot(sc.keys(), sc.values())
plt.title('Silhouette score plot')
plt.xlabel('Cluster')
plt.ylabel('Silhouette score')
```

```python
# from Elbow plot and Silhouette score plot we can conclude that the optimum
# number of cluster is 3.
kmeans = KMeans(n_clusters=3, random_state = 0).fit(scaled_df)
rfm_data_copy['Clusters'] = kmeans.predict(scaled_df)
```

```python
print(rfm_data_copy)
```

```python
# Cluster Analysis
rfm_data_copy['Clusters'].value_counts()
```

```python
c0 = (rfm_data_copy['Clusters'].value_counts()[0] / len(rfm_data_copy)) * 100
c1 = (rfm_data_copy['Clusters'].value_counts()[1] / len(rfm_data_copy)) * 100
c2 = (rfm_data_copy['Clusters'].value_counts()[2] / len(rfm_data_copy)) * 100
print('Cluster 0: ', str(c0) + ' %')
print('Cluster 1: ', str(c1) + ' %')
print('Cluster 2: ', str(c2) + ' %')
```

```python
rfm_data_copy.drop(columns=['CustomerID'], inplace = True)
mean = rfm_data_copy.groupby('Clusters').mean()
median = rfm_data_copy.groupby('Clusters').median()
df_kmeans = pd.concat([mean, median], axis = 0)
df_kmeans.index = ['0 mean', '1 mean', '2 mean', '0 median', '1 median', '2 median']
df_kmeans.T
```

## Cluster Analysis Report
- Cluster 2 customers - These are the customers who **purchase in large quantity and visits the store frequently**. This is the cluster which has low number of elements (**0.27 %**) means that we can expect this kind of customers rarely.
- Cluster 1 customers - These are the customers who **visits shop very frequently but purchase occasionally**. They will buy a low amount of purchase and visit the shop with very high frequency compare to other two clusters. This cluster has about **25% of the total customers**.
- Cluster 0 customers - These are the customers who **actively purchase products and visits the shop quite often**. These are the customers who visits the shop with a decent frequency and purchase a decent number of products every time. About **74% of customers lies in this cluster.**
